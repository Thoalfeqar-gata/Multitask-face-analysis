# Splitting data into training and testing using the 80/20 split.

# AgeDB

```python
import os
import numpy as np
import pandas as pd
import shutil
```

```python
lst = [0, 32, 32, 32]
indices = np.cumsum(lst)
indices
```

```python
path = os.path.join('data', 'datasets', 'age gender and race estimation', 'AgeDB')
```

```python
files = os.listdir(path)
files = [file for file in files if file.endswith(".jpg")]
indices = np.arange(start = 0, stop = len(files))
```

```python
np.random.shuffle(indices)
```

```python
os.makedirs(name = os.path.join(path, 'test'), exist_ok = True)
```

```python
os.makedirs(name = os.path.join(path, 'train'), exist_ok = True)
```

```python
indices_split = int(0.8 * len(indices))
train_indices = indices[:indices_split]
test_indices = indices[indices_split:]
```

```python
for index in train_indices:
    shutil.move(src = os.path.join(path, files[index]), dst = os.path.join(path, 'train', files[index]))
```

```python
for index in test_indices:
    shutil.move(src = os.path.join(path, files[index]), dst = os.path.join(path, 'test', files[index]))
```
