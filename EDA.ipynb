# Exploratory Data Analysis

## Face recognition datasets

### Glint360k

```python
import os, matplotlib.pyplot as plt, numpy as np
from utility_scripts.data_loaders import VGGFace_dataset, MS1MV2_dataset, CasiaWebFace_dataset, Glint360k_dataset
```

```python
full_vgg_dataset = Glint360k_dataset(
    dataset_dir = os.path.join('data', 'datasets', 'face recognition', 'glint360k'),
    image_transform = None,
    target_transform = None,
    subset = None,
)
```

```python
from pympler import asizeof
```

```python
asizeof.asizeof(full_vgg_dataset) / (1024**3)  # Size in GB
```

```python
# show some samples
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
for i in range(3):
    for j in range(3):
        idx = np.random.randint(0, len(full_vgg_dataset))
        img, label = full_vgg_dataset[idx]
        axes[i, j].imshow(img.permute(1, 2, 0))  # Assuming img is a tensor with shape (C, H, W)
        axes[i, j].set_title(f'Label: {label}')
        axes[i, j].axis('off')
plt.tight_layout()
plt.show()
```
