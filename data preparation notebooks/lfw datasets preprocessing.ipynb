```python
import os
os.chdir('..')
from matplotlib import pyplot as plt
```

```python
import torch, PIL, numpy as np, cv2
from adaface_utils.face_alignment.align import get_aligned_face
```

```python
datasets = [
    os.path.join('LFW', 'lfw-deepfunneled'), 
    os.path.join('cplfw', 'aligned images'), 
    os.path.join('calfw', 'aligned images')
]

base_dir = os.path.join('data', 'datasets', 'face recognition')
```

# LFW

```python
complete_lfw_path = os.path.join(base_dir, datasets[0])
output_lfw_path = os.path.join(base_dir, os.path.split(datasets[0])[0], 'processed images')
```

```python
for dir in os.listdir(complete_lfw_path):
    output_path = os.path.join(output_lfw_path, dir)
    
    if not os.path.exists(output_path):
        os.mkdir(output_path)

    detector_failed = False
    for file in os.listdir(os.path.join(complete_lfw_path, dir)):

        image_path = os.path.join(complete_lfw_path, dir, file)
        pil_image = PIL.Image.open(image_path)

        aligned_image = get_aligned_face(image_path = None, rgb_pil_image = pil_image)
        if aligned_image is not None:
            aligned_image = np.array(aligned_image)[:, :, ::-1] #convert to bgr so cv2 stores it properly
        else:
            #just center crop if the face detector doesn't detect the image
            aligned_image = np.array(pil_image)[:, :, ::-1]
            aligned_image = aligned_image[(250-112)//2:-(250-112)//2, (250-112)//2:-(250-112)//2] #all images are original of size 250x250, so center crop to 112x112
            alinged_image = aligned_image[:, :, ::-1]
        
        aligned_image = cv2.resize(aligned_image, (112, 112))
        cv2.imwrite(os.path.join(output_path, file), aligned_image)

  
```
