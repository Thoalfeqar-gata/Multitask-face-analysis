# Extracting CFP_FP and CFP_FF

```python
import pickle, numpy as np, os
from PIL import Image
import io
import cv2
import pandas as pd
```

## CFP_FP

```python
path_cfp_fp = os.path.join('data', 'datasets', 'face recognition', 'cfp_fp.bin')
```

```python
with open(path_cfp_fp, 'rb') as f:
    bins, issame_list = pickle.load(f, encoding = 'bytes')
```

Each two consecutive images are a pair, 
and the label indicates same or not (verification)

```python
cfp_fp_output_dir = os.path.join('data', 'datasets', 'face recognition', 'cfp_fp')
```

```python
for i in range(len(bins)):
    image = cv2.imdecode(
        np.frombuffer(bins[i], dtype = np.uint8),
        cv2.IMREAD_COLOR
    )

    cv2.imwrite(os.path.join(cfp_fp_output_dir, f'{i}.jpg'), image)
```

```python
pair_1 = [f'{i}.jpg' for i in range(0, len(bins), 2)]
pair_2 = [f'{i+1}.jpg' for i in range(0, len(bins), 2)]
```

```python
df = pd.DataFrame([pair_1, pair_2, issame_list]).T
```

```python
df.to_csv(os.path.join(
    'data', 'datasets', 'face recognition', 'cfp_fp', 'labels.csv'
), index = False, header = False)
```

# CFP_FF

```python
path_cfp_ff = os.path.join('data', 'datasets', 'face recognition', 'cfp_ff.bin')
```

```python
cfp_ff_output_dir = os.path.join('data', 'datasets', 'face recognition', 'cfp_ff')
```

```python
with open(path_cfp_ff, 'rb') as f:
    bins, issame_list = pickle.load(f, encoding = 'bytes')
```

```python
for i in range(len(bins)):
    image = cv2.imdecode(
        np.frombuffer(bins[i], dtype = np.uint8),
        cv2.IMREAD_COLOR
    )

    cv2.imwrite(os.path.join(cfp_ff_output_dir, 'images', f'{i}.jpg'), image)
```

```python
pair_1 = [f'{i}.jpg' for i in range(0, len(bins), 2)]
pair_2 = [f'{i+1}.jpg' for i in range(0, len(bins), 2)]
```

```python
df = pd.DataFrame([pair_1, pair_2, issame_list]).T
```

```python
df.to_csv(os.path.join(
    'data', 'datasets', 'face recognition', 'cfp_ff', 'labels.csv'
), index = False, header = False)
```
